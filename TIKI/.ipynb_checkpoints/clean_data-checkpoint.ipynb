```python
# Thay đổi giá trị category1_name
product['category1_name'] = product['category1_name'].replace(
    {'Áo sơ mi nữ': 'Áo sơ mi - Áo kiểu', 'Áo kiểu nữ': 'Áo sơ mi - Áo kiểu'}
)
```

```python
# Đếm số lượng sản phẩm trong từng nhóm `category2_id`
category_counts = product['category2_id'].value_counts()

# Lọc ra các `category2_id` có số lượng sản phẩm >= 10
valid_categories = category_counts[category_counts >= 10].index

# Giữ lại các dòng có `category2_id` nằm trong danh sách hợp lệ
product = product[product['category2_id'].isin(valid_categories)]
```

```python
product['category2_name'].nunique()
```

```python
product['category2_name'].value_counts()
```

```python
# Tìm và thay thế các category1_name chứa cụm từ "áo thun nữ tay ngắn..."
product['category2_name'] = product['category2_name'].replace(
    {'Áo thun nữ ngắn tay không cổ ': 'Áo thun tay ngắn', 'Áo thun nữ ngắn tay có cổ': 'Áo thun tay ngắn',
     'Áo thun nữ dài tay không cổ ': 'Áo thun dài tay', 'Áo thun nữ dài tay có cổ': 'Áo thun dài tay'}
)
```

```python
# Đếm số lượng sản phẩm trong từng nhóm `category2_id`
category2_counts = product['category2_name'].value_counts()

# Lọc ra các `category2_id` có số lượng sản phẩm >= 10
valid_categories2 = category2_counts[category2_counts >= 10].index

# Giữ lại các dòng có `category2_id` nằm trong danh sách hợp lệ
product = product[product['category2_name'].isin(valid_categories2)]
```

```python
categories_to_split = [
    'Áo thun tay ngắn',
    'Đầm suông',
    'Áo sơ mi nữ tay dài',
    'Đầm Dáng Xòe',
    'Quần tây nữ',
    'Áo len - Áo khoác nỉ nữ',
    'Áo kiểu nữ khác',
    'Chân váy dài'
]

# Khởi tạo DataFrame chứa các dòng cần giữ lại
df_kept = pd.DataFrame()

# Xử lý từng loại
for category in categories_to_split:
    # Lấy tất cả các dòng thuộc loại này
    df_category = product[product['category2_name'] == category]
    
    # Chọn ngẫu nhiên 50% dòng để giữ lại
    df_sampled = df_category.sample(frac=0.5, random_state=42)
    
    # Thêm vào DataFrame kết quả
    df_kept = pd.concat([df_kept, df_sampled])

# Thêm các dòng không thuộc danh sách cần xử lý
df_other = product[~product['category2_name'].isin(categories_to_split)]

# Gộp tất cả các dòng lại
product = pd.concat([df_kept, df_other])
```

```python
from bs4 import BeautifulSoup
import re
def clean_html(description):
    if pd.isna(description):  # Kiểm tra nếu là giá trị NaN
        return None
    soup = BeautifulSoup(description, "html.parser")
    text = soup.get_text()
    text_cleaned = re.sub(r'http\S+', '', text)  # Loại bỏ URL
    return text_cleaned

# Áp dụng hàm cho toàn bộ cột 'description'
product['description'] = product['description'].apply(clean_html)
```
